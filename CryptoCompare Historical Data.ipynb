```python
cryptocompare.get_historical_price('BTC', 'EUR', datetime.datetime(2020,11,11))
```

```python
daily_btc= cryptocompare.get_historical_price_day('BTC', curr='USD')
daily_btc
```

```python
daily_btc_data = pd.DataFrame(daily_btc, columns = ['time','open','high','low','close'])
daily_btc_data
```

```python
daily_btc_data['20 SMA'] = daily_btc_data.close.rolling(20).mean()
daily_btc_data.tail()
```

```python
daily_btc_data['time'] = pd.to_datetime(daily_btc_data['time'], unit='s')
daily_btc_data.tail()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(daily_btc_data.time, daily_btc_data.close)
plt.xticks(rotation=45)
plt.show()
```
