```python
url = "https://min-api.cryptocompare.com/data/index/histo/underlying/hour"

payload = {
    "market" : "CCMVDA",
    "base" : "BTC",
    "quote" : "USD"
    
}

result = requests.get(url, headers=headers, params=payload).json()
result['Data']
```

```python
df=pd.DataFrame(result['Data'], columns=['time','open','high','low','close'])
df['time'] = pd.to_datetime(df['time'], unit='s')
plt.plot(df.time, df.close)
plt.xticks(rotation=45)
plt.show()
df.set_index('time',inplace=True)
df.head()
```
