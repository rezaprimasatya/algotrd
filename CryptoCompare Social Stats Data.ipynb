```python
url = "https://min-api.cryptocompare.com/data/social/coin/histo/day"

hist_day = requests.get(url, headers=headers).json()
hist_day['Data']
```

```python
dfs = pd.DataFrame(hist_day['Data'], columns=['time','comments','posts','followers'])
dfs['time'] = pd.to_datetime(dfs['time'], unit='s')
dfs.tail()
```

```python
plt.plot(dfs.time, dfs.followers)
plt.xticks(rotation=45)
plt.show()
```
